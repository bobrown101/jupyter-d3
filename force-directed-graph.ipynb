The code below will load d3.js with require.js

If you don't know what require js does, all you need to know is it manages dependencies, and we need to hook into it in order for our page to correctly access the d3 module

```python
%%javascript
require.config({
  paths: {
    d3: "http://d3js.org/d3.v3.min"
  }
});

require(["d3"], function(d3) {
//   d3.select("body").prepend("h1").text("Successfully loaded D3 version " + d3.version);
    alert("Loaded d3");
    console.log("%c%s",
            "color: red; background: yellow; font-size: 24px;",
            "Successfully loaded d3:");
    console.log(d3);
});
```
