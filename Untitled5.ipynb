```python
from IPython.display import clear_output
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
import d3_lib
import random
```

```python
HTML(
"""
<script type="text/javascript" src="http://d3js.org/d3.v3.min.js"></script>
<script>
   console.log(d3);
</script>
""")
```

```python

num_nodes = 30
num_edges = 0.05

slider_nodes = widgets.IntSlider()
slider_edges = widgets.IntSlider()
submit_button = widgets.Button(description="Create Graph")


def slider_nodes_handler(change):
    num_nodes = change['new']
    
def slider_edges_handler(change):
    num_edges = change['new']

    
def on_button_click(change):
    print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
    display_graph(num_nodes, num_edges)
    print("=========================================================")
    
slider_nodes.observe(slider_nodes_handler, names='value')
slider_edges.observe(slider_edges_handler, names='value')

submit_button.on_click(on_button_click)


def display_graph(num_nodes, num_edges): 
    graph = {"nodes": [], "links": []}
    for i in range(num_nodes):
        graph["nodes"].append( {"name": "i" + str(i), "group": int(random.uniform(1,100))} )
    for i in range(num_nodes):
        for j in range(num_nodes):
            if random.uniform(0,1) < num_edges:
                graph["links"].append( {"source": i, "target": j, "value": random.uniform(1,30)} )
    clear_output()
    display(HTML(d3_lib.set_styles(['force_directed_graph'])))
    display(HTML(d3_lib.draw_graph('force_directed_graph',{'data': graph})))
    
    

    
```
